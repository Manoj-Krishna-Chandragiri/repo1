```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# 1️⃣ Load and preprocess MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 2️⃣ Normalize images (convert pixels to range [0,1])
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape to match CNN input format (28x28 grayscale images)
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# 3️⃣ Convert labels to categorical (10 classes: digits 0-9)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 4️⃣ Build CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')  # 10 output classes (digits 0-9)
])

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 5️⃣ Train with varying epochs and store accuracy
epochs_list = [5, 10, 15]
accuracy_list = []

for epochs in epochs_list:
    print(f"\nTraining model with {epochs} epochs...\n")
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=64, validation_data=(X_test, y_test), verbose=1)

    # Evaluate model
    _, accuracy = model.evaluate(X_test, y_test)
    accuracy_list.append(accuracy)

# 6️⃣ Plot accuracy vs. epochs
plt.plot(epochs_list, accuracy_list, marker='o', linestyle='-')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('CNN Accuracy vs. Epochs')
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16, VGG19
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# 1️⃣ Load and Preprocess Dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values (0 to 1)
X_train, X_test = X_train / 255.0, X_test / 255.0

# Convert labels to categorical format (one-hot encoding for 10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2️⃣ Load Pre-Trained VGG16 Model (without top layers)
base_vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Freeze pre-trained layers (they are already trained)
for layer in base_vgg16.layers:
    layer.trainable = False

# Add custom layers for CIFAR-10 classification
x = Flatten()(base_vgg16.output)
x = Dense(256, activation='relu')(x)
x = Dense(10, activation='softmax')(x)  # 10 output classes

# Define the final VGG16 model
vgg16_model = Model(inputs=base_vgg16.input, outputs=x)

# Compile the model
vgg16_model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Train VGG16
print("\nTraining VGG16...\n")
history_vgg16 = vgg16_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))

# Evaluate VGG16
_, vgg16_acc = vgg16_model.evaluate(X_test, y_test)
print(f"VGG16 Accuracy: {vgg16_acc:.4f}")

# 3️⃣ Load Pre-Trained VGG19 Model (without top layers)
base_vgg19 = VGG19(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Freeze pre-trained layers
for layer in base_vgg19.layers:
    layer.trainable = False

# Add custom layers for CIFAR-10 classification
y = Flatten()(base_vgg19.output)
y = Dense(256, activation='relu')(y)
y = Dense(10, activation='softmax')(y)  # 10 output classes

# Define the final VGG19 model
vgg19_model = Model(inputs=base_vgg19.input, outputs=y)

# Compile the model
vgg19_model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Train VGG19
print("\nTraining VGG19...\n")
history_vgg19 = vgg19_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))

# Evaluate VGG19
_, vgg19_acc = vgg19_model.evaluate(X_test, y_test)
print(f"VGG19 Accuracy: {vgg19_acc:.4f}")

# 4️⃣ Compare Accuracy and Plot Results
epochs = range(1, 11)
plt.plot(epochs, history_vgg16.history['val_accuracy'], label="VGG16 Accuracy")
plt.plot(epochs, history_vgg19.history['val_accuracy'], label="VGG19 Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.title('VGG16 vs VGG19 Accuracy')
plt.legend()
plt.show()
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess the data
X_train = X_train.reshape(-1, 28 * 28).astype('float32') / 255
X_test = X_test.reshape(-1, 28 * 28).astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
model = Sequential([
    Dense(128, activation='relu', input_shape=(28 * 28,)),
    Dropout(0.2),  # Dropout layer with 20% dropout rate
    Dense(64, activation='relu'),
    Dropout(0.2),  # Another Dropout layer with 20% dropout rate
    Dense(10, activation='softmax')  # Output layer with 10 neurons (one for each class)
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Plot training and validation loss to show the effect of dropout
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python

import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize pixel values between 0 and 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Load VGG16 without the top layer (fully connected layers)
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Freeze base model layers
for layer in base_model.layers:
    layer.trainable = False

# Create the model
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(256, activation='relu'))   # Custom dense layer
model.add(Dropout(0.5))                    # Dropout for regularization
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(10, activation='softmax')) # Output layer for 10 classes

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_test, y_test))

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python

import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout

# 1️⃣ Load the IMDB dataset
num_words = 10000  # Keep the most frequent 10,000 words
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words)

# 2️⃣ Pad sequences to make all reviews the same length
max_length = 200  # Maximum number of words per review
X_train = pad_sequences(X_train, maxlen=max_length, padding='post')
X_test = pad_sequences(X_test, maxlen=max_length, padding='post')

# 3️⃣ Build the LSTM model
model = Sequential([
    Embedding(input_dim=num_words, output_dim=128, input_length=max_length),  # Word embeddings
    LSTM(64, return_sequences=True),  # First LSTM layer
    Dropout(0.2),  # Prevent overfitting
    LSTM(32),  # Second LSTM layer
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# 4️⃣ Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 5️⃣ Train the model
model.fit(X_train, y_train, epochs=5, batch_size=64, validation_data=(X_test, y_test))

# 6️⃣ Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical

# 1️⃣ Load Dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# 2️⃣ Normalize pixel values (0-1 range)
X_train, X_test = X_train / 255.0, X_test / 255.0

# 3️⃣ Convert labels to categorical format (10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 4️⃣ Define the Neural Network Model
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Convert 2D image to 1D array
    Dense(128, activation='relu'),  # Hidden layer with 128 neurons
    Dense(64, activation='relu'),   # Hidden layer with 64 neurons
    Dense(10, activation='softmax') # Output layer (10 classes)
])

# 5️⃣ Compile the Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 6️⃣ Implement Early Stopping
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# 7️⃣ Train the Model with Early Stopping
history = model.fit(X_train, y_train, epochs=50, batch_size=64,
                    validation_data=(X_test, y_test),
                    callbacks=[early_stop])

# 8️⃣ Evaluate Model
_, accuracy = model.evaluate(X_test, y_test)
print(f"Final Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
# Sample text data
text = """deep learning is amazing and deep learning models are powerful.
machine learning is the future and artificial intelligence is evolving.
deep learning and artificial intelligence are changing the world."""

# Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# Convert words to numerical sequences
total_words = len(tokenizer.word_index) + 1  # Add 1 for padding
sequences = []
for sentence in text.split('.'):
    words = sentence.strip().split()
    for i in range(1, len(words)):
        sequence = words[:i+1]
        sequences.append(sequence)

# Convert sequences to numerical format
input_sequences = []
for seq in sequences:
    input_sequences.append([tokenizer.word_index[word] for word in seq])

# Pad sequences to equal length
max_length = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_length, padding='pre')

# Split into inputs (X) and labels (y)
X, y = input_sequences[:, :-1], input_sequences[:, -1]
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
model = Sequential([
    Embedding(total_words, 100, input_length=max_length-1),  # Word embeddings
    LSTM(150, return_sequences=True),  # LSTM layer
    LSTM(100),  # Another LSTM layer
    Dense(100, activation='relu'),  # Fully connected layer
    Dense(total_words, activation='softmax')  # Output layer
])

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=100, verbose=1)
def predict_next_word(seed_text, tokenizer, max_length):
    sequence = tokenizer.texts_to_sequences([seed_text])[0]
    sequence = pad_sequences([sequence], maxlen=max_length-1, padding='pre')
    predicted_index = np.argmax(model.predict(sequence, verbose=0), axis=-1)
    return tokenizer.index_word.get(predicted_index[0], "unknown")

# Example Prediction
seed_text = "deep learning"
next_word = predict_next_word(seed_text, tokenizer, max_length)
print(f"Next word prediction: {next_word}")
```
